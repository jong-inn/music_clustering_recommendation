## __packages__

```python
import os, json, unicodedata
import pandas as pd

pickle_path_tracks = "../sampled_data/sampled_data_tracks_final.pickle"
pickle_path_track_artist = "../sampled_data/sampled_data_track_artist_final.pickle"
raw_path_artist = "../extracted_data/artists/artists_df.pickle"
```

## __unicode function__

```python
def unicode_normalize(string):
    
    normalized = unicodedata.normalize("NFD", string)
    return u"".join([c for c in normalized if not unicodedata.combining(c)])
```

## __load data__

```python
artists = pd.read_pickle(raw_path_artist)
tracks = pd.read_pickle(pickle_path_tracks)
track_artist = pd.read_pickle(pickle_path_track_artist)

artists["name"] = artists.name.apply(unicode_normalize)
tracks["track_name"] = tracks.track_name.apply(unicode_normalize)
track_artist["artist_name"] = track_artist.artist_name.apply(unicode_normalize)

display(artists.head())
display(tracks.head())
display(track_artist.head())
```

## __extract sample data__

```python
tmp_tracks = tracks.iloc[:1000, :]
tmp_track_artist = track_artist.loc[track_artist.track_id.isin(tmp_tracks.track_id)]

print(tmp_tracks.shape)
display(tmp_tracks.head())

print(tmp_track_artist.shape)
display(tmp_track_artist.head())
```

```python
artist_col = ["artist_id", "name", "genres", "popularity"]
sample_artist = artists.loc[artists.artist_id.isin(tmp_track_artist.artist_id.unique())]
sample_artist = sample_artist[artist_col]

album_col = ["album_id", "album_popularity", "release_date"]
sample_album = tmp_tracks.loc[:, album_col]
sample_album.drop_duplicates(subset="album_id", inplace=True)

track_col = ["track_id", "album_id", "track_name", "duration_ms"]
sample_track = tmp_tracks.loc[:, track_col]
sample_track["popularity"] = 0

audio_features_col = ["tempo", "valence", "danceability", "acousticness", "energy", "instrumentalness", "liveness", "loudness", "speechiness", "track_id"]
sample_audio_features = tmp_tracks.loc[:, audio_features_col]
sample_audio_features["pseudo_id"] = range(0,len(sample_audio_features))
sample_audio_features = sample_audio_features[["pseudo_id"] + audio_features_col]

sample_artist_album = pd.merge(tmp_track_artist.set_index("track_id")[["artist_id"]], tmp_tracks.set_index("track_id")[["album_id"]], how="left", left_index=True, right_index=True)
sample_artist_album.reset_index(inplace=True, drop=True)
sample_artist_album.drop_duplicates(inplace=True)

sample_artist_track = tmp_track_artist[["artist_id", "track_id"]]

print(sample_artist.shape)
display(sample_artist.head())

print(sample_album.shape)
display(sample_album.head())

print(sample_track.shape)
display(sample_track.head())

print(sample_audio_features.shape)
display(sample_audio_features.head())

print(sample_artist_album.shape)
display(sample_artist_album.head())

print(sample_artist_track.shape)
display(sample_artist_track.head())
```
